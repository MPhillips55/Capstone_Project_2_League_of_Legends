### Reshaping the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
from tsfresh import defaults
%matplotlib inline
```

```python
df = pd.read_csv('gold_diff_merged.csv', index_col=0)
```

```python
df.drop(['league','season', 'year', 'team_name', 'game_length', 'player_name', 'champion'], inplace=True, axis=1)
```

```python
cols = ['match_id', 'red_or_blue_side', 'result', 'type']
df.rename(columns=lambda x: x[4:] if x not in cols else x,inplace=True)
df.head()
```

```python
# annotate match_id to differentiate the roles/games
df['match_id2'] = df['match_id']
df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'total_gold'),'match_id2'] = df['match_id2'] + '_b'
df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'total_gold'),'match_id2'] = df['match_id2'] + '_r'

df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'top_gold'),'match_id2'] = df['match_id2'] + '_b'
df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'jungle_gold'),'match_id2'] = df['match_id2'] + '_b'
df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'mid_gold'),'match_id2'] = df['match_id2'] + '_b'
df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'adc_gold'),'match_id2'] = df['match_id2'] + '_b'
df.loc[(df['red_or_blue_side'] == 'blue') & (df['type'] == 'support_gold'),'match_id2'] = df['match_id2'] + '_b'

df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'top_gold'),'match_id2'] = df['match_id2'] + '_r'
df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'jungle_gold'),'match_id2'] = df['match_id2'] + '_r'
df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'mid_gold'),'match_id2'] = df['match_id2'] + '_r'
df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'adc_gold'),'match_id2'] = df['match_id2'] + '_r'
df.loc[(df['red_or_blue_side'] == 'red') & (df['type'] == 'support_gold'),'match_id2'] = df['match_id2'] + '_r'
```

```python
m_id = df['match_id2']
df.drop(['match_id','match_id2'], axis=1, inplace=True)
df.insert(0, 'match_id', m_id)
df.head()
```

```python
# separate the data into df's for each role and total
tdf = df.loc[(df['type'] == 'total_gold'),:].copy()
topdf = df.loc[(df['type'] == 'top_gold'),:].copy()
jdf = df.loc[(df['type'] == 'jungle_gold'),:].copy()
mdf = df.loc[(df['type'] == 'mid_gold'),:].copy()
adf = df.loc[(df['type'] == 'adc_gold'),:].copy()
sdf = df.loc[(df['type'] == 'support_gold'),:].copy()
```

```python
# drop extraneous columns
tdf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
topdf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
jdf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
mdf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
adf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
sdf.drop(['type', 'red_or_blue_side','result'], axis=1,inplace=True)
```

```python
# reshape using melt 
# make sure the 'minute' column is cast as int
tdf2 = pd.melt(tdf, id_vars=['match_id'])
tdf2.variable = tdf2.variable.astype(int)

topdf2 = pd.melt(topdf, id_vars=['match_id'])
topdf2.variable = topdf2.variable.astype(int)

jdf2 = pd.melt(jdf, id_vars=['match_id'])
jdf2.variable = jdf2.variable.astype(int)

mdf2 = pd.melt(mdf, id_vars=['match_id'])
mdf2.variable = mdf2.variable.astype(int)

adf2 = pd.melt(adf, id_vars=['match_id'])
adf2.variable = adf2.variable.astype(int)

sdf2 = pd.melt(sdf, id_vars=['match_id'])
sdf2.variable = sdf2.variable.astype(int)
```

```python
# sort after the merge
tdf1 = tdf2.sort_values(['match_id', 'variable'])
topdf1 = topdf2.sort_values(['match_id', 'variable'])
jdf1 = jdf2.sort_values(['match_id', 'variable'])
mdf1 = mdf2.sort_values(['match_id', 'variable'])
adf1 = adf2.sort_values(['match_id', 'variable'])
sdf1 = sdf2.sort_values(['match_id', 'variable'])
```

```python
# sanity check
print(len(tdf1))
print(len(topdf1))
print(len(jdf1))
print(len(mdf1))
print(len(adf1))
print(len(sdf1))
```

```python
tdf1.reset_index(drop=True, inplace=True)
topdf1.reset_index(drop=True, inplace=True)
jdf1.reset_index(drop=True, inplace=True)
mdf1.reset_index(drop=True, inplace=True)
adf1.reset_index(drop=True, inplace=True)
sdf1.reset_index(drop=True, inplace=True)
```

```python
# rename columns to make the final df readable
tdf1.rename(columns={'variable':'minute','value':'gold_diff'},inplace=True)
topdf1.rename(columns={'variable':'minute','value':'top_gold_diff'},inplace=True)
jdf1.rename(columns={'variable':'minute','value':'jgl_gold_diff'},inplace=True)
mdf1.rename(columns={'variable':'minute','value':'mid_gold_diff'},inplace=True)
adf1.rename(columns={'variable':'minute','value':'adc_gold_diff'},inplace=True)
sdf1.rename(columns={'variable':'minute','value':'sup_gold_diff'},inplace=True)
sdf1.head()
```

```python
# begin merging all the melted df's to one
tdf3 = pd.merge(tdf1, topdf1, on=['match_id','minute'])
tdf3.head()
```

```python
tdf3 = pd.merge(tdf3, jdf1, on=['match_id','minute'])
tdf3.head()
```

```python
tdf3 = pd.merge(tdf3, mdf1, on=['match_id','minute'])
tdf3.head()
```

```python
tdf3 = pd.merge(tdf3, adf1, on=['match_id','minute'])
tdf3.head()
```

```python
tdf3 = pd.merge(tdf3, sdf1, on=['match_id','minute'])
tdf3.head()
```

```python
len(tdf3)
```

```python
# the data includes empty rows (0's) for each minute up to 80, even if the game ended
# the following drop will remove the extra rows, but keep the valid 0's that occur early in a game
tdf3 = tdf3.drop(tdf3[(tdf3.minute > 10) & (tdf3.gold_diff == 0) & (tdf3.top_gold_diff == 0) & 
                      (tdf3.jgl_gold_diff == 0) & (tdf3.mid_gold_diff == 0) & (tdf3.adc_gold_diff == 0)
                     & (tdf3.sup_gold_diff == 0)].index)
```

```python
len(tdf3)
```

```python
tdf3.reset_index(drop=True,inplace=True)
```

```python
# create 'y' df
tdf = df.loc[(df['type'] == 'total_gold'),:].copy()
y = tdf[['match_id', 'result']]
print(len(y))
y.set_index('match_id', inplace=True)
y = y['result']
y.head()
```

```python
from tsfresh import extract_relevant_features

features = extract_relevant_features(tdf3, y, column_id='match_id', column_sort='minute')
```
